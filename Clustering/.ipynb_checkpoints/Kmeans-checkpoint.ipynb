```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

data, labels = load_digits(return_X_y=True)
(n_samples, n_features), n_digits = data.shape, np.unique(labels).size

print(f"# digits: {n_digits}; # samples: {n_samples}; # features {n_features}")
# print(data)
# plt.imshow(data[1000].reshape(8,8))
# print(labels[1000])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=n_digits, n_init=4, random_state=0)

kmeans.fit(X_train)
y_pred = kmeans.predict(X_test)
# print(y_pred)
# print(y_test)
```

```python
from sklearn.metrics import rand_score
rs = rand_score(y_test, y_pred)
print("rand_score:", rs)
```

```python
centroids = kmeans.cluster_centers_
print(centroids.shape)
import matplotlib.pyplot as plt

centroids_viz = scaler.inverse_transform(centroids)
# print(centroids.shape)

for i in range(centroids_viz.shape[0]):
    fig = plt.figure()
    plt.imshow(centroids_viz[i].reshape(8,8))
```
